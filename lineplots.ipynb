```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('honeyproduction.csv')
dfCA = df.loc[lambda df: df['state'] == 'CA', :]
dfND = df.loc[lambda df: df['state'] == 'ND', :]
dfSD = df.loc[lambda df: df['state'] == 'SD', :]


#reminder that if I use an integer I get a type error because of an invalid comparison. Python reads 'integer' as a string while it's an integer in the df.
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.scatter(dfCA['year'], dfCA['numcol'], label='California', color='gold')
plt.scatter(dfCA['year'], dfND['numcol'], label='North Dakota', color='silver')
plt.scatter(dfCA['year'], dfSD['numcol'], label='South Dakota', color='black')
#plt.xlim([1998, 2012])
#I cannot change the limits of the x-axis without ruinging the graph. found a workaround by using a 2-D line plot
plt.xlabel('Year')
plt.ylabel('Number of Honeybee Colonies')
plt.title('Number of Honeybee colonies during 1998-2012')
plt.legend()

plt.figure(figsize=(12,5))
plt.subplot(1,2,2)
plt.plot(dfCA['year'], dfCA['yieldpercol'], label='California', color='gold')
plt.plot(dfCA['year'], dfND['yieldpercol'], label='North Dakota', color='silver')
plt.plot(dfCA['year'], dfSD['yieldpercol'], label='South Dakota', color='black')
plt.xlabel('Year')
plt.ylabel('Yield per Colony (lbs)')
plt.title('Yield per Honeybee colony 1998-2012')
plt.legend()
#odd text shows up in my output.

```
