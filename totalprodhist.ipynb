```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('honeyproduction.csv')
dfCA = df.loc[lambda df: df['state'] == 'CA', :]
dfND = df.loc[lambda df: df['state'] == 'ND', :]
dfSD = df.loc[lambda df: df['state'] == 'SD', :]

#plt.hist(df['priceperlb'], bins=50)
plt.hist(dfCA['totalprod'], color='red', bins=20, alpha=0.5, label='California')
plt.hist(dfND['totalprod'], color='blue', bins=20, alpha=0.5, label='North Dakota')
plt.hist(dfSD['totalprod'], color='yellow', bins=20, alpha=0.5, label='South Dakota')
plt.title('Total Honey Production 1998-2012')
plt.xlabel('Pounds of Honey (x 10^7lbs)')
plt.legend()
```
