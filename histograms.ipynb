```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

#importing libraries
```

```python
CBK = pd.read_csv('BK 17_19.csv')
CAST= pd.read_csv('ASTORIA 17_19.csv')
CJH = pd.read_csv('JH 17_19.csv')

ABK = pd.read_csv('BK JUNE_19.csv')
AST = pd.read_csv('ASTORIA JUNE_19.csv')
AJH = pd.read_csv('JH JUNE_19.csv')

#loading files
```

```python
CBK.drop([0,1], inplace = True)
CAST.drop([0,1], inplace = True)
CJH.drop([0,1], inplace = True)
ABK.drop([0,1], inplace = True)

# Removing extra rows with no info

```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 6))
fig.suptitle('Distribution of Risk Scores Closed Cases 2017-2019', fontsize=16)
axs[0].set_ylabel('Number of Cases', Fontsize=14)

axs[0].hist(CBK['Risk Level'], bins=20)
axs[0].set_xlabel('Brooklyn',Fontsize=14);

axs[1].hist(CAST['Risk Level'], bins=20)
axs[1].set_xlabel('Astoria',Fontsize=14);

axs[2].hist(CJH['Risk Level'], bins=20);
axs[2].set_xlabel('Jackson Heights',Fontsize=14);

#First set of Histograms
```

```python
plt.figure(figsize=(18,7)) 


plt.subplot(1, 3, 1)
ax = CBK.groupby('Language').size().plot(kind='bar');
plt.xticks(rotation=45);
plt.xlabel('Brooklyn',Fontsize=14);
plt.ylabel('Number of Instances',Fontsize=14);
plt.locator_params(axis='y', nbins=40)


plt.subplot(1, 3, 2)
ax = CAST.groupby('Language').size().plot(kind='bar');
plt.title('Distribution of Languages Closed Cases 2017-2019', fontsize=16);
plt.xticks(rotation=45);
plt.xlabel('Astoria',Fontsize=14);
plt.locator_params(axis='y', nbins=40)


plt.subplot(1, 3, 3)
ax = CJH.groupby('Language').size().plot(kind='bar');
plt.xticks(rotation=45);
plt.xlabel('Jacson Heights',Fontsize=14);
plt.locator_params(axis='y', nbins=40)



```

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 7))
fig.suptitle('Distribution of Risk Scores for Active Cases as of 6-19-19', fontsize=16)
axs[0].set_ylabel('Number of Cases', Fontsize=14)

axs[0].hist(ABK['Risk Level'], bins=20)
axs[0].set_xlabel('Brooklyn',Fontsize=14);

axs[1].hist(AST['Risk Level'], bins=20)
axs[1].set_xlabel('Astoria',Fontsize=14);

axs[2].hist(AJH['Risk Level'], bins=20);
axs[2].set_xlabel('Jackson Heights',Fontsize=14);
```
