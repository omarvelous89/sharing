```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('honeyproduction.csv')
dfCA = df.loc[lambda df: df['state'] == 'CA', :]
dfND = df.loc[lambda df: df['state'] == 'ND', :]
dfSD = df.loc[lambda df: df['state'] == 'SD', :]

plt.plot(dfCA['priceperlb'], dfCA['prodvalue'], label='California')
plt.plot(dfND['priceperlb'], dfND['prodvalue'], label='North Dakota')
plt.plot(dfSD['priceperlb'], dfSD['prodvalue'], label='South Dakota')
plt.title('Total Product Value .vs. Price per lb of Honey')
plt.xlabel('Price per lb ($)')
plt.ylabel('Total Product Value ($)')
plt.legend()
```
