```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('honeyproduction.csv')
dfCA = df.loc[lambda df: df['state'] == 'CA', :]
dfND = df.loc[lambda df: df['state'] == 'ND', :]
dfSD = df.loc[lambda df: df['state'] == 'SD', :]

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(dfCA['year'], dfCA['prodvalue'], label='California')
plt.plot(dfND['year'], dfND['prodvalue'], label='North Dakota')
plt.plot(dfSD['year'], dfSD['prodvalue'], label='South Dakota')
plt.title('Total Product Value .vs. Year')
plt.xlabel('Year')
plt.ylabel('Total Product Value ($)')
plt.legend()

plt.subplot(1,2,2)
plt.plot(dfCA['year'], dfCA['stocks'], label='California')
plt.plot(dfND['year'], dfND['stocks'], label='North Dakota')
plt.plot(dfSD['year'], dfSD['stocks'], label='South Dakota')
plt.title('Stocks in Honey 1998-2012')
plt.xlabel('year')
plt.ylabel('stocks owned')
plt.legend()
```
